```python
import torch
import torchvision
import numpy as np
import pandas as pd
from PIL import Image, ImageDraw
from torch.utils.data import Dataset
from torchvision import transforms as T
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
from sklearn.model_selection import train_test_split
```

```python
PATH_TO_DATASET="./drive/MyDrive/Flowers_ML/DatasetPic/"
IMAGE_TYPE=".jpg"
```

```python
def collate_fn(data):
    return data
```

```python
class CustData(Dataset):
    def __init__(self, image_name, bounding_boxes):
        self.image_paths = image_name
        self.bounding_boxes = bounding_boxes

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, index):
        image_path = PATH_TO_DATASET + self.image_paths.iloc[index] + IMAGE_TYPE
        boxes = self.bounding_boxes.iloc[index].to_numpy().astype(float)

        # Загрузка изображения
        image = Image.open(image_path).convert('RGB')
        # Инициализцаия трансформера
        transformer = T.ToTensor()
        # Трансформ изображения
        image_tensor = transformer(image)
        boxes = torch.as_tensor(boxes, dtype=torch.float32).unsqueeze(0)
        # there is only one class
        labels = torch.ones((1), dtype=torch.int64)
        target = {}
        target["boxes"] = boxes
        target["labels"] = labels

        # Возвращаем данные в формате, удобном для обучения модели
        return image_tensor, target
```

```python
def model_init():
    model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained = True)
    num_classes = 2
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)
    return model

def param_init(model):
    optimizer = torch.optim.SGD(model.parameters(), lr=0.005, momentum=0.9, weight_decay=0.0005)
    lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer,
                                                   step_size=3,
                                                   gamma=0.1)
    num_epochs = 10
    return num_epochs, optimizer, lr_scheduler
```

```python
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

train_csv = pd.read_excel('./drive/MyDrive/Flowers_ML/BB.xlsx')
train_csv.head()

df = train_csv['bbox'].str.split(',', expand=True)
df.rename(columns={0:'x0', 1:'y0', 2:'x1', 3:'y1'}, inplace=True)

df = pd.concat([train_csv, df], axis=1)
del df['bbox']

train_inds, val_inds = train_test_split(range(len(df)), test_size=0.1)

dataset_train = CustData(df['name'].iloc[train_inds], df[['x0','y0','x1','y1']].iloc[train_inds])
dataset_val = CustData(df['name'].iloc[val_inds], df[['x0','y0','x1','y1']].iloc[val_inds])
# разделим на трейн и валидацию
data_loader = torch.utils.data.DataLoader(
    dataset_train, batch_size=2, shuffle=True,
    collate_fn=collate_fn,
    )

data_loader_test = torch.utils.data.DataLoader(
    dataset_val, batch_size=1, shuffle=False,
    collate_fn=collate_fn)

model = model_init().to(device)
num_epochs, optimizer, lr_scheduler = param_init(model)

for epoch in range(num_epochs):
    epoch_loss = 0
    for data_batch in data_loader:
        img = []
        targets = []
        for data in data_batch:
            img.append(data[0].to(device))
            targ = {}
            targ["boxes"] = data[1]["boxes"].to(device)
            targ["labels"] = data[1]["labels"].to(device)
            targets.append(targ)
        loss_dict = model(img, targets)
        loss = sum(v for v in loss_dict.values())
        epoch_loss += loss.cpu().detach().numpy()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(epoch_loss)

model.eval()
for data_test in data_loader_test:
    img = data_test[0][0]
    output = model([img.to(device)])
out_bbox = output[0]["boxes"]
out_scores = output[0]["scores"]
keep = torchvision.ops.nms(out_bbox, out_scores, 0.5)
out_bbox = out_bbox[keep]
out_scores = out_scores[keep]
im = (img.permute(1,2,0).cpu().detach().numpy() * 255).astype('uint8')
vsample = Image.fromarray(im)
draw = ImageDraw.Draw(vsample)
for box in out_bbox:
    draw.rectangle(list(box), fill= None, outline='red')
vsample
```
